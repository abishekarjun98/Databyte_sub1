```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
df=pd.read_csv("company_sales_data.csv")
```

```
df.head()
```

```
plt.plot(df["month_number"],df["total_profit"],color="red")
plt.xticks(np.arange(1,13,1)) # to get each month's number on the x axis
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.title("Month vs Total Profit")
```

```
x=df["month_number"]
plt.figure(figsize=(7,7))
for column in df.iloc[:,1:7]: #iterating through various cloumns of the dataset
  y=df[column]
  print(column)
  plt.xticks(np.arange(1,13,1))
  plt.plot(x,y,label=column)
plt.legend()  
plt.xlabel("Month")
plt.ylabel("Profits of various Products")
plt.title("Product Sales Data")
#data points of facewash and moisturizer are same, so only 5 plots are visible
```

```

Totalsales=[]#a list which hold sales data for a year
Products=[] 
net_sales_of_a_product=0
for column in df.iloc[:,1:7]:
  for sales in df[column]:
    net_sales_of_a_product+=sales #adding up sales of a particular product to get net sales for an year
  Totalsales.append(net_sales_of_a_product) 
  Products.append(column)
  net_sales_of_a_product=0 
print(Products)
print(Totalsales)


fig1, ax1 = plt.subplots()
ax1.pie(Totalsales, labels=Products, autopct='%1.1f%%') #having various products as labels and Totalsales as data

ax1.axis('equal') 

plt.title("Total Sales of Products in a Year")
plt.show()

    
```
